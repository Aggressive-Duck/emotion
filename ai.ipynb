```python
#text preprocessing (clean Data)
import pandas as pd
df=pd.read_excel("./content/Emotion616.xlsx")
#drop the columns
df=df.drop(columns=['Sr. no.'])
df.head()
```

```python
#convert to lowercase
df['clean_text']=df["Emails"].str.lower()
df.head()
```

```python
#remove of punctuations(標點符號)
import string
print(string.punctuation)
```

```python
def remove_punctuation(text):
  return text.translate(str.maketrans('','',string.punctuation))
```

```python
df['clean_text']=df['clean_text'].apply(lambda x:remove_punctuation(x))
df.head()
```

```python
def replace_space(text):
  return text.replace("\n"," ")
```

```python
df['clean_text']=df['clean_text'].apply(lambda x:replace_space(x))
df.head()
```

```python
#remove of stopwords
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
STOPWORDS=set(stopwords.words('english'))
def remove_stopword(text):
  return ' '.join([word for word in text.split() if word not in STOPWORDS])
```

```python
df['clean_text']=df['clean_text'].apply(lambda x:remove_stopword(x))
df.head()
```

```python
#remove frequenct words
from collections import Counter
cnt=Counter()
for text in df['clean_text']:
  for word in text.split():
    cnt[word]+=1
cnt.most_common(10)
```

```python
FREQUENT_WORDS = set([word for (word, wc) in cnt.most_common(10)])
def remove_frequent_words(text):
  return ' '.join([word for word in text.split() if word not in FREQUENT_WORDS])
```

```python
df['clean_text']=df['clean_text'].apply(lambda x:remove_frequent_words(x))
df.head()
```

```python
#remove rare words
RARE_WORDS = set([word for (word, wc) in cnt.most_common()[:-10:-1]])
def remove_rare_words(text):
  return ' '.join([word for word in text.split() if word not in RARE_WORDS])
```

```python
df['clean_text']=df['clean_text'].apply(lambda x:remove_rare_words(x))
df.head()
```

```python
from nltk import pos_tag
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer

nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger_eng')

lemmatizer = WordNetLemmatizer()
word_map = {"N":wordnet.NOUN, "V":wordnet.VERB, "J":wordnet.ADJ, "R":wordnet.ADV}
def lemmatize_words(text):
  pos_text = pos_tag(text.split())
  return ' '.join([lemmatizer.lemmatize(word, word_map.get(pos[0], wordnet.NOUN)) for word, pos in pos_text])
```

```python
df['lemmatized_text'] = df['clean_text'].apply(lambda x:lemmatize_words(x))
df.head()
```

```python
df['lemmatized_text'][0]
#原句子:Respected sir,The beauty of the Northern areas and your effort made the people to visit these beautiful areas.
```

```python
#in-text features(word count(字數),punctuation(表點符號),sentiment(情緒),emoticons(表情符號))
```

```python
#第1步:先把文字轉成數值 ex:Email text:"I love dogs" ----->[0.2, 0, 0.5, 0.1, …] 過程叫做:Vectorization(向量化)
#第2步:Scale Features:數值歸一化（Normalization / Standardization）
#第3步:PCA 降維
```

```python
df['lemmatized_text'].shape#預處理完後目前339筆資料
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorize=TfidfVectorizer(max_features=100)#整個資料集挑出最重要的100個詞
x=vectorize.fit_transform(df['lemmatized_text']).toarray()#fit根據學習詞表做訓練 transform轉成數字向量
print(x.shape)#已轉換成每一筆資料有100維 有339筆資料
print("標準化前")
print(x[0])
```

```python
#正規化
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
print("標準化後")
print(x_scaled[0])
```

```python
#PCA
from sklearn.decomposition import PCA
pca=PCA(n_components=0.95)#「100 維 → 壓縮成 81 維，但資訊只損失 5%」
x_pca=pca.fit_transform(x_scaled)
print(x_pca.shape)
print(x_pca[0])

```

```python
#SVM (只接受數字標籤ex:0,1,2...) 所以要先把六種情緒轉換成相對應的數字
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y_encode=le.fit_transform(df['Label'])
print(y_encode)
print(le.classes_)
#Happy:0
#angry=1
#negatively suprised=2
#neutral=3
#positively surprised=4
#sad=5
```

```python
#分割資料集
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
    x_pca, y_encode, test_size=0.2, random_state=42
)
```

```python
#kernel="rbf" 使用非線性 能將原本線性分不開的資料，映射到高維度，使其能被分開 (升維轉換是一種核技巧)
#C=1 模型容忍錯誤 越小間隔(margin)較寬 允許錯誤 防止過擬合
#gamma='scale'控制每個訓練樣本對決策邊界的影響範圍。 gamma小 每個樣本影響decision boundary 範圍大 變界平滑 scale:自動計算
#decision_function_shape='ovr' 'ovr' → One-vs-Rest 每次訓練「這一類 vs 其他類」，共訓練 6 個二分類模型。另一種是OVO 每兩類之間訓練一個二分類器
```

```python
#建立支持向量機
from sklearn.svm import SVC #SVC是SVM的分類版本 Support Vector Classifier
svm = SVC(kernel='rbf', C=1, gamma='scale', decision_function_shape='ovr')
svm.fit(x_train,y_train)
```

```python
#預測
y_pred=svm.predict(x_test)
```

```python
#評估模型
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test,y_pred)
print("Accuracy:",acc)
```
